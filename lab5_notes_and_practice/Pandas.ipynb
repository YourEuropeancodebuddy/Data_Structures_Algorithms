Accessing data for instance a column  from the data frame we pass the column as an index 
- retrieving a single column provides a data series 
- retrieving multiple columns give a tabular arrangment 

```python
moviereview
moviereview[["movies", "duration", "rating"]] # the order of the colum is according to the order provided 
# and not the original data 
```

- accessing the rows the key word `loc` is used

```python
moviereview.loc[0]
# comes as a series with the columns as index 
```

```python
# to recover specific ones a list of the index is passed
moviereview.loc[[2,4]]
```

```python
# to recover specific rows and specific columns

moviereview.loc[1:4, ["movies", "duration","rating"]]
```

```python
# to select one element the at can be used
moviereview.at[1, "duration"]
```
